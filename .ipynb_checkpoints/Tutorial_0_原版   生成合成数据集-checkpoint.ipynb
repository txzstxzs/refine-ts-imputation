# Diffusion-TS Tutorial

## Diffusion-TS: Interpretable Diffusion for General Time Series Generation

This notebook describes the user-guide of a time-series synthetic data generation application using Diffusion-TS framework.

## Necessary packages and functions call

```python
'-----sine数据集设置epoch 7000往上时pca和tsne效果较好 几乎完全重合 虽然设置7000以下等值也可以较早收敛 最终loss值也差不多 但是pca等效果没有那么好-----'

import os
import torch
import numpy as np

from engine.solver import Trainer
from Utils.metric_utils import visualization
from Data.build_dataloader import build_dataloader
from Utils.io_utils import load_yaml_config, instantiate_from_config
from Models.interpretable_diffusion.model_utils import unnormalize_to_zero_to_one
```

## Setting and Training

### Build dataset and settings

# 选择数据集的配置文件 对于sins mujoco合成数据集 可以在配置里改参数 如改序列长度 特征数等

```python
class Args_Example:
    def __init__(self) -> None:
        self.config_path = './Config/sines.yaml'     # 选择数据集
#         self.config_path = './Config/stocks.yaml'
#         self.config_path = './Config/Energy.yaml'
#         self.config_path = './Config/fmri.yaml'
#         self.config_path = './Config/MuJoCo.yaml'  # 需要dm_control
        self.save_dir = './toy_exp'
        self.gpu = 0
        os.makedirs(self.save_dir, exist_ok=True)

args =  Args_Example()
configs = load_yaml_config(args.config_path)
device = torch.device(f'cuda:{args.gpu}' if torch.cuda.is_available() else 'cpu')


configs['solver']['max_epochs']=5000
configs['dataloader']['batch_size'] = 128
print(configs['model'])
print(configs['solver'])
print(configs['dataloader']['train_dataset'])
print(configs['dataloader']['test_dataset'])
print(configs['dataloader']['batch_size'])
print(configs['dataloader']['sample_size'])
```

```python
dl_info = build_dataloader(configs, args)   # 此时会把真实数据保存为npy格式
dl = dl_info['dataloader']
print(next(iter(dl)).shape)

ds = dl_info['dataset']
print(len(ds), ds[0].shape)

# 

# def cycle(dl):
#     while True:
#         for data in dl:
#             yield data
# cycle(dl)

# for data in dl:
#     print(data.shape)
```

```python
model = instantiate_from_config(configs['model']).to(device)   # 导入Diffusion_TS
# print(model)

trainer = Trainer(config=configs, args=args, model=model, dataloader=dl_info)
```

### Training models

```python
trainer.train()
```

## Inference and Visualizations

### Sampling

```python
dataset = dl_info['dataset']
seq_length, feature_dim = dataset.window, dataset.var_num
print(seq_length, feature_dim, dataset.dir, len(dataset), dataset.auto_norm )
```

```python
# ori_data = np.load(os.path.join( dataset.dir, f"sine_ground_truth_{seq_length}_train.npy" ))

ori_data = np.load(os.path.join(dataset.dir, f"{'stock'}_norm_truth_{seq_length}_train.npy"))
print(ori_data.shape)
```

```python
fake_data = trainer.sample( num=len(dataset), size_every=500, shape=[seq_length, feature_dim]) # 采样同样大小的生成数据 size_every是每次生成的个数
print(fake_data.shape)
```

```python
'这个反归一化对pca有影响 要使用 '
'只能用一次 不然pca效果变差 因为生成数据不对了'
if dataset.auto_norm:
    fake_data = unnormalize_to_zero_to_one(fake_data)
    np.save(os.path.join(args.save_dir, f'ddpm_fake_sines.npy'), fake_data)
```

### Visualization

We visualize the original and synthetic data distributions using PCA, tSNE and Density analysis.

```python
visualization(ori_data=ori_data, generated_data=fake_data, analysis='pca', compare=ori_data.shape[0])
```

```python
visualization(ori_data=ori_data, generated_data=fake_data, analysis='tsne', compare=ori_data.shape[0])
```

```python
visualization(ori_data=ori_data, generated_data=fake_data, analysis='kernel', compare=ori_data.shape[0])
```
