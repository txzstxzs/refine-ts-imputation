```python
def parse_args():
    parser = argparse.ArgumentParser(description='PyTorch Training Script')
    parser.add_argument('--name', type=str, default=None)

    parser.add_argument('--config_file', type=str, default=None, 
                        help='path of config file')
    parser.add_argument('--output', type=str, default='OUTPUT', 
                        help='directory to save the results')
    parser.add_argument('--tensorboard', action='store_true', 
                        help='use tensorboard for logging')

    # args for random

    parser.add_argument('--cudnn_deterministic', action='store_true', default=False,
                        help='set cudnn.deterministic True')
    parser.add_argument('--seed', type=int, default=10, 
                        help='seed for initializing training.')
    parser.add_argument('--gpu', type=int, default=None,
                        help='GPU id to use. If given, only the specific gpu will be'
                        ' used, and ddp will be disabled')
    
    # args for training
    parser.add_argument('--train', action='store_true', default=False, help='Train or Test.')
    parser.add_argument('--sample', type=int, default=0, 
                        choices=[0, 1], help='Condition or Uncondition.')
    parser.add_argument('--mode', type=str, default='infill',
                        help='Infilling or Forecasting.')
    parser.add_argument('--milestone', type=int, default=10)

    parser.add_argument('--missing_ratio', type=float, default=0., help='Ratio of Missing Values.')
    parser.add_argument('--pred_len', type=int, default=0, help='Length of Predictions.')
    
    
    # args for modify config
    parser.add_argument('opts', help='Modify config options using the command-line',
                        default=None, nargs=argparse.REMAINDER)  

    args = parser.parse_args(args=[])
    args.save_dir = os.path.join(args.output, f'{args.name}')

    return args
```

```python
import torch
from torch.utils.data import Dataset
from torch.utils.data import DataLoader

if data_name == 'SP500':
    load_data = np.load( './Data/datasets/sp500.npy')

elif data_name == 'ETTh1':
    load_data = np.load( './Data/datasets/ETTh1.npy')
    
elif data_name == 'Energy':
    load_data = np.load( './Data/datasets/Energy.npy')

elif data_name == 'MuJoCo':
    load_data = np.load( './Data/datasets/MuJoCo.npy')
    
    
# np.random.seed(11)
np.random.shuffle(load_data)
train_data = load_data[ 0: int( load_data.shape[0]*0.8) ]
test_data = load_data[ int( load_data.shape[0]*0.8): ]    

    
print(train_data.shape, test_data.shape)
# print(train_data[0])

Numble = train_data.shape[0]
Length = train_data.shape[1]
Feature = train_data.shape[2]
Batchsize = 128


class MyDataset(Dataset):
    def __init__(self, X_data):
        self.X_data = X_data

    def __len__(self):
        return len(self.X_data)

    def __getitem__(self, idx):
        x = torch.tensor(self.X_data[idx], dtype=torch.float32)  # 转换为 Tensor
        return x   
                    
dataset = MyDataset(train_data)    
dataloader = DataLoader(dataset, batch_size=128, shuffle=True)

def cycle(dl):         
    while True:
        for data in dl:
            yield data
            
dl = cycle( dataloader )
print( next(dl).shape )
```

```python
device = model.betas.device
step = 0

'这个要10000  5000不够好 但是loss几乎一样'
train_num_steps = 5000
gradient_accumulate_every = config['solver']['gradient_accumulate_every'] 
```

```python
start_lr = config['solver'].get('base_lr', 1.0e-4)
ema_decay = config['solver']['ema']['decay']
ema_update_every = config['solver']['ema']['update_interval']
ema = EMA(model, beta=ema_decay, update_every=ema_update_every).to(device)

opt = Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=start_lr, betas=[0.9, 0.96])
# print(opt)

sc_cfg = config['solver']['scheduler']
sc_cfg['params']['optimizer'] = opt    # 配置里添加了一个优化器参数
sch = instantiate_from_config(sc_cfg)   # 导入了 engine.lr_sch.ReduceLROnPlateauWithWarmup
# print(sch)
```

```python
loss_list = []

pbar = tqdm( total=train_num_steps )   

while step < train_num_steps:
    total_loss = 0.
    
    for _ in range(gradient_accumulate_every):    # 2  循环2次 然后损失要分半  这一部分是否用好像影响不大
        
        data = next(dl).to(device)           # [128, 24, 6]  前面对dl用了cycle 类似iter 可以取数据
        loss = model(data, target=data)        # 计算损失 可以加上频域损失
        
        loss_list.append(loss.item())

        loss = loss / gradient_accumulate_every
        loss.backward()                      # 梯度反传了 但是还没有用step 模型还没更新 后面要剪裁
        total_loss += loss.item()
        
        
    clip_grad_norm_(model.parameters(), 1.0)   # 剪裁参数 再优化器step  新建的优化器也step
    opt.step()
    sch.step(total_loss)    # 这一步影响最大  不用这个的话 loss收敛有问题 且波动大
    opt.zero_grad()
    step += 1
    ema.update()          # 这个影响好像不大
       
    pbar.update(1)
    
pbar.close()
```

```python
'把采样方法分离开来'

from Models.interpretable_diffusion.model_utils import default, identity, extract
from Models.interpretable_diffusion.gaussian_diffusion import linear_beta_schedule, cosine_beta_schedule, Diffusion_TS
import torch.nn.functional as F


shape = (Numble, Length, Feature)
print(shape)

num_timesteps = model.num_timesteps
print(num_timesteps)


if beta_schedule == 'linear':
    betas = linear_beta_schedule(num_timesteps)
elif beta_schedule == 'cosine':
    betas = cosine_beta_schedule(num_timesteps)
            
        
alphas = 1. - betas
alphas_cumprod = torch.cumprod(alphas, dim=0)
alphas_cumprod_prev = F.pad(alphas_cumprod[:-1], (1, 0), value=1.)
print(betas.shape)

print(betas[0:10])
print(betas[0:10].to(torch.float32))
```

```python
betas = betas.to(torch.float32).cuda()
alphas = alphas.to(torch.float32).cuda()
alphas_cumprod = alphas_cumprod.to(torch.float32).cuda()
alphas_cumprod_prev = alphas_cumprod_prev.to(torch.float32).cuda()

# calculations for diffusion q(x_t | x_{t-1}) and others
sqrt_alphas_cumprod = torch.sqrt(alphas_cumprod).to(torch.float32).cuda()
sqrt_one_minus_alphas_cumprod = torch.sqrt(1. - alphas_cumprod).to(torch.float32).cuda()
log_one_minus_alphas_cumprod = torch.log(1. - alphas_cumprod).to(torch.float32).cuda()
sqrt_recip_alphas_cumprod =  torch.sqrt(1. / alphas_cumprod).to(torch.float32).cuda()
sqrt_recipm1_alphas_cumprod =  torch.sqrt(1. / alphas_cumprod - 1).to(torch.float32).cuda()


# calculations for posterior q(x_{t-1} | x_t, x_0)
posterior_variance = betas * (1. - alphas_cumprod_prev) / (1. - alphas_cumprod).cuda()

# above: equal to 1. / (1. / (1. - alpha_cumprod_tm1) + alpha_t / beta_t)
posterior_variance = posterior_variance.to(torch.float32).cuda()

posterior_log_variance_clipped = torch.log(posterior_variance.clamp(min=1e-20)).to(torch.float32).cuda()
posterior_mean_coef1 = betas * torch.sqrt(alphas_cumprod_prev) / (1. - alphas_cumprod).to(torch.float32).cuda()
posterior_mean_coef2 =  (1. - alphas_cumprod_prev) * torch.sqrt(alphas) / (1. - alphas_cumprod).to(torch.float32).cuda()

# calculate reweighting
loss_weight =  ( torch.sqrt(alphas) * torch.sqrt(1. - alphas_cumprod) / betas / 100).to(torch.float32).cuda()


print(sqrt_alphas_cumprod.shape, posterior_variance.shape )
```

```python
'DDPM采样'
'把下面三个参数去掉的话影响都不大'
padding_masks = None
clip_x_start = False
clip_denoised = True


'简单来看 就是先用xt预测x0 再计算xt-1'
with torch.no_grad():     
    
    x = torch.randn(shape, device=device)
    
    for t in tqdm( reversed(range(0, num_timesteps)) ):
        
        ori_t = t
        batched_times = torch.full((x.shape[0],), t, device=x.device, dtype=torch.long)  # 扩散步加维度
        t = batched_times
        
        '这个掩码对结果影响好像不大 好像是注意力模块里的一个参数'
        if padding_masks is None:
            padding_masks = torch.ones(x.shape[0], Length, dtype=bool, device=x.device)

            
        '这个好像影响不大  identity就是返回输入值'
        maybe_clip = partial(torch.clamp, min=-1., max=1.) if clip_x_start else identity
        
        
        '预测x0 表示为趋势和周期'  
        trend, season = ema.ema_model.model(x, t, padding_masks=padding_masks)     # 输入xt 预测x0  用趋势和周期表示       
        x_start = trend + season 
        x_start = maybe_clip(x_start)
        
        
        '根据x0计算噪声  这个噪声后面没有用到'
#         pred_noise = ( extract(sqrt_recip_alphas_cumprod, t, x.shape ) * x - x_start) /  extract(sqrt_recipm1_alphas_cumprod, t, x.shape)  
             
        '限制数据幅值 限制再-1和1之间 使用mujoco数据集时不用这个 因为该数据集好像没有归一化过 限制幅度的话结果就不对了'
        if clip_denoised:
            x_start.clamp_(-1., 1.)
        
        '根据x0 xt 计算xt-1的均值方差'
        mean = ( extract( posterior_mean_coef1, t, x.shape) * x_start +
                           extract( posterior_mean_coef2, t, x.shape) * x )

        variance = extract( posterior_variance, t, x.shape)  # 方差应该是固定的
        log_variance_clipped = extract( posterior_log_variance_clipped, t, x.shape)
        
        
        '计算xt-1'
        noise = torch.randn_like(x) if ori_t > 0 else 0.                     # no noise if t == 0
        x = mean + (0.5 * log_variance_clipped).exp() * noise   # 计算xt-1  均值加标准差 右边化简后就是标准差
    
    
fake_data = x.detach().cpu().numpy()
print(fake_data.shape)
```

```python
'DDIM采样'

padding_masks = None
clip_x_start = False
clip_denoised = True
shape = (Numble, Length, Feature)
eta = 0
batch = shape[0]
total_timesteps = num_timesteps     # 总的扩散步
sampling_timesteps = num_timesteps  # 快速采样的步数 可以小于总的扩散步数从而加速采样
                                    # 降到50以下效果也不错

with torch.no_grad(): 
    
    # ddim采样 根据采样步 把扩散步分成等间隔的步数   采样步和扩散步相同时[-1, 0, 1, 2, ..., T-1] when sampling_timesteps == total_timesteps
    times = torch.linspace(-1, total_timesteps - 1, steps=sampling_timesteps + 1)   
    times = list(reversed(times.int().tolist()))
    time_pairs = list(zip(times[:-1], times[1:]))  # [(T-1, T-2), (T-2, T-3), ..., (1, 0), (0, -1)]
    
    x = torch.randn(shape, device=device)
    
    
    for time, time_next in tqdm(time_pairs, desc='sampling loop'):
        
        batched_times = torch.full((batch,), time, device=device, dtype=torch.long)
        t = batched_times
        
        if padding_masks is None:
            padding_masks = torch.ones(x.shape[0], Length, dtype=bool, device=x.device)
            
        maybe_clip = partial(torch.clamp, min=-1., max=1.) if clip_x_start else identity
        
        
        '预测x0 表示为趋势和周期'  
        trend, season = ema.ema_model.model(x, t, padding_masks=padding_masks)     # 输入xt 预测x0  用趋势和周期表示       
        x_start = trend + season             
        x_start = maybe_clip(x_start)
        
        'x0和xt的转换公式 计算噪声'
        pred_noise = ( extract(sqrt_recip_alphas_cumprod, t, x.shape ) * x - x_start) /  extract(sqrt_recipm1_alphas_cumprod, t, x.shape)  
              
        if time_next < 0:
            x = x_start     # 最后一步预测值
            continue

        'DDIM的多个系数 计算xt-1  这里计算时sigma不为0 也就不是确定性采样'
        alpha = alphas_cumprod[time]
        alpha_next = alphas_cumprod[time_next]
        sigma = eta * ((1 - alpha / alpha_next) * (1 - alpha_next) / (1 - alpha)).sqrt()
        c = (1 - alpha_next - sigma ** 2).sqrt()
        noise = torch.randn_like(x)
        
        x = x_start * alpha_next.sqrt() +  c * pred_noise +  sigma * noise

    
fake_data = x.detach().cpu().numpy()
print(fake_data.shape)
```

```python
def visualization2(ori, gen):
    
    Number = ori.shape[0]
    
    max_sample = min([4000, Number])       # 最多取4000个样本
    idx = np.random.permutation(Number)[ :max_sample]

    ori1 = np.asarray(ori)[idx]
    gen1 = np.asarray(gen)[idx]

    ori1 = np.mean(ori1, 2) 
    gen1 = np.mean(gen1, 2) 


    '上部分为原样本 下部分为生成样本'
    concat_data = np.concatenate((ori1, gen1), axis=0)
    
    pca = PCA(2)     
    pca_res = pca.fit_transform(concat_data)

    tsne = TSNE(n_components=2, verbose=1, perplexity=40,random_state=20,n_iter=500) 
    tsne_res = tsne.fit_transform(concat_data) 

    umap_ = umap.UMAP(random_state=42)
    umap_res = umap_.fit_transform(concat_data)
    
    
    fig = plt.figure(figsize=(19,4))
    plt.subplot(1,4,1)
    
    '0到2000索引的为原样本'
    plt.scatter(pca_res[:2000,0],pca_res[:2000,1], s=15, c='#ff7f0e')
    plt.scatter(pca_res[max_sample:max_sample+2000,0],pca_res[max_sample:max_sample+2000,1], s=15, c='#1f77b4')
    plt.xlabel('(a)',fontsize=25)
    
    plt.subplot(1,4,2)
    plt.scatter(tsne_res[:2000,0],tsne_res[:2000,1],  s=15,  c='#ff7f0e')
    plt.scatter(tsne_res[max_sample:max_sample+2000,0],tsne_res[max_sample:max_sample+2000,1], s=15, c='#1f77b4')
    plt.xlabel('(b)',fontsize=25)

    plt.subplot(1,4,3)
    plt.scatter(umap_res[:2000, 0], umap_res[:2000, 1],  cmap='Spectral', s=15,  c='#ff7f0e')
    plt.scatter(umap_res[max_sample:max_sample+2000, 0], umap_res[max_sample:max_sample+2000, 1],  cmap='Spectral', s=15, c='#1f77b4')
    plt.xlabel('(c)',fontsize=25)

    
    plt.subplot(1,4,4)
    sns.distplot(ori[:,:,1], hist=False, color = '#ff7f0e' )
    sns.distplot(gen[:,:,1], hist=False, color = '#1f77b4' )  
    plt.xlabel('(d)',fontsize=25)
    
    
#     plt.savefig('pca等合成图1111',dpi=500, bbox_inches='tight' )
#     plt.savefig('pca等合成图_sp5001111.pdf', bbox_inches='tight')
```

```python
'DPS采样  需要掩码数据集'

mr = 0.3


'使用自带的掩码 这里面好像没有mnr掩码'

from Utils.masking_utils import noise_mask


'这个方法做出的掩码每个都不相同 用true false表示 后面封装为dataset时也可以转换为01'
'separate是mcar 另外的不清楚是什么掩码 '
'geometric应该是每个掩码都不一样 否则应该是一样 '
def mask_data( data, missing_ratio, mean_mask_length=3, style='separate', distribution='geometric', seed=2023): 
        masks = np.ones_like( data )
        # Store the state of the RNG to restore later.
        st0 = np.random.get_state()
#         np.random.seed(seed)

        for idx in range(data.shape[0]):
            x = data[idx, :, :]  # (seq_length, feat_dim) array
            mask = noise_mask(x, missing_ratio, mean_mask_length, style, distribution)  # (seq_length, feat_dim) boolean array
            masks[idx, :, :] = mask

        # Restore RNG.
        np.random.set_state(st0)
        return masks.astype(bool)
#         return masks
    
        
mask_data = mask_data(data=test_data, missing_ratio=mr)
print(mask_data.shape)
# print(mask_data[0][0:30])
print( np.sum(mask_data[0]==0)/ (Length*Feature)  )
```

```python
'使用sssd里的mnr掩码'
import random

mr = 0.3


def get_mask_mcar(sample, mr): 
    mask = torch.ones(sample.shape).view(1,-1) 
    index = torch.tensor(range(mask.shape[1]))  
    perm = torch.randperm(len(index)) 
    index = perm[ 0: round( mr*len(index) ) ]
    mask[0][index]=0
    mask = mask.view(sample.shape)
    return mask


def get_mask_mnr(sample, k):  # 缺失的时候是连续k个点一起缺失 不那么随机 
    """Get mask of random segments (non-missing at random) across channels based on k,
    where k == number of segments. Mask of sample's shape where 0's to be imputed, and 1's to preserved
    as per ts imputers"""

    mask = torch.ones(sample.shape)
    length_index = torch.tensor(range(mask.shape[0]))
    list_of_segments_index = torch.split(length_index, k)
    for channel in range(mask.shape[1]):
        s_nan = random.choice(list_of_segments_index)
        mask[:, channel][s_nan[0]:s_nan[-1] + 1] = 0

    return mask




missing_length = int( mr * Length )
mask = torch.empty([0, Length, Feature])
        
for i in range(load_data.shape[0]):
    one_mask = get_mask_mcar( train_data[0], mr)  
#     one_mask = get_mask_mnr( train_data[0], missing_length)   
    one_mask = one_mask.unsqueeze(0)
    mask = torch.row_stack([mask, one_mask])

mask_data = mask.numpy().astype(bool)
print(mask_data.shape)
# print(mask_data[0][0:30])
print( np.sum(mask_data[0]==0)/ (Length*Feature)  )
```

```python
'DPS采样'
'基本过程是先计算xt-1(ddpm或ddim) 然后计算观测一致性的梯度 微调xt-1'
import time

coef = 1.0e-2
coef_=0.
step_size = 5.0e-2
learning_rate = step_size
sampling_steps =  num_timesteps
shape = (Length, Feature)
padding_masks = None
clip_x_start = False
clip_denoised = True

#  空文件 用于保存结果时的初始文件
samples = np.empty([0, shape[0], shape[1]])
reals = np.empty([0, shape[0], shape[1]])
masks = np.empty([0, shape[0], shape[1]])    


with torch.no_grad(): 
    
    for idx, (x, t_m) in  enumerate(dataloader_with_mask):     # 取出数据和掩码 掩码用true false表示
        x, t_m = x.to(device), t_m.to(device)
        partial_mask = t_m
        target = x*t_m

        xt = torch.randn(x.shape, device=device)   # 默认数据类型 torch.float32 要和dataloader里的x的数据类型一致 在建立dataloader时可以修改数据类型

        
        start_time = time.time()
        
        '先计算xt-1 再用dps微调'
        for t in tqdm(reversed(range(0, num_timesteps)), desc='DPS loop'):

            ori_t = t
            batched_times = torch.full((xt.shape[0],), t, device=x.device, dtype=torch.long)
            t = batched_times

            if padding_masks is None:
                padding_masks = torch.ones(x.shape[0], Length, dtype=bool, device=x.device)

            maybe_clip = partial(torch.clamp, min=-1., max=1.) if clip_x_start else identity

            trend, season = ema.ema_model.model(xt, t, padding_masks=padding_masks)   # 根据xt开始计算x0 再计算xt-1的均值方差 
            x_start = trend + season 
            x_start = maybe_clip(x_start)

#             pred_noise = ( extract(sqrt_recip_alphas_cumprod, t, xt.shape ) * xt - x_start) /  extract(sqrt_recipm1_alphas_cumprod, t, xt.shape) 

            if clip_denoised:
                x_start.clamp_(-1., 1.)

            '计算xt-1的均值方差'
            mean = ( extract( posterior_mean_coef1, t, xt.shape) * x_start +
                     extract( posterior_mean_coef2, t, xt.shape) * xt )

            variance = extract( posterior_variance, t, xt.shape)  # 方差应该是固定的
            log_variance_clipped = extract( posterior_log_variance_clipped, t, xt.shape)


            '计算xt-1'
            noise = torch.randn_like(xt) if ori_t > 0 else 0.        # no noise if t == 0
            sigma = (0.5 * log_variance_clipped).exp()
            xt = mean + sigma * noise                              # 计算xt-1  均值加标准差 右边化简后就是标准差



            '根据DPS开始计算需要调整的梯度 微调xt-1'
            if t[0].item() < num_timesteps * 0.05:   # k是dps里用梯度更新xt-1的次数  
                K = 0
            elif t[0].item() > num_timesteps * 0.9:
                K = 3
            elif t[0].item() > num_timesteps * 0.75:
                K = 2
                learning_rate = learning_rate * 0.5   # 学习率在后面逐步减小
            else:
                K = 1
                learning_rate = learning_rate * 0.25

            input_embs_param = torch.nn.Parameter(xt)   # 把xt-1看成可更新参数 用观测一致性损失来调整

        
            with torch.enable_grad():              # 由于要更新xt-1 所以下面要传递梯度
                for i in range(K):                 # dps的更新次数
                    optimizer = torch.optim.Adagrad([input_embs_param], lr=learning_rate)  # xt-1的优化器
                    optimizer.zero_grad()

                    trend, season = ema.ema_model.model(input_embs_param, t)  # 输入xt-1预测x0  和观测值计算观测一致性损失
                    x_start = trend + season

                    # 计算dps的系数和观测一致性损失 微调xt-1
                    if sigma.mean() == 0:
                        logp_term = coef * ((mean - input_embs_param) ** 2 / 1.).mean(dim=0).sum()
                        infill_loss = (x_start[partial_mask] - target[partial_mask]) ** 2        # 观测处损失
                        infill_loss = infill_loss.mean(dim=0).sum()
                    else:
                        logp_term = coef * ((mean - input_embs_param)**2 / sigma).mean(dim=0).sum()  
                        infill_loss = (x_start[partial_mask] - target[partial_mask]) ** 2       # 观测一致性损失
                        infill_loss = (infill_loss/sigma.mean()).mean(dim=0).sum()

                    loss = logp_term + infill_loss
                    loss.backward()
                    optimizer.step()
                    epsilon = torch.randn_like(input_embs_param.data)
                    input_embs_param = torch.nn.Parameter((input_embs_param.data + coef_ * sigma.mean().item() * epsilon).detach()) # 用损失的梯度更新xt-1


            xt[~partial_mask] = input_embs_param.data[~partial_mask]   # xt-1的缺失部分一致 观测部分保留


            # 观测位置加噪
#             noise = lambda: torch.randn_like(target)
            noise = torch.randn_like(target)
            target_t =  extract(sqrt_alphas_cumprod, t, target.shape) * target +  \
                        extract(sqrt_one_minus_alphas_cumprod, t, target.shape) * noise 
        
           
            xt[partial_mask] = target_t[partial_mask]              # xt-1的观测位置也加噪 方便下一个采样步的计算 此时缺失位置已经用dps更新

            
        end_time = time.time()
        execution_time = end_time - start_time   
        print(execution_time)  

        
        '注意 此时只是一个batch的样本处理好 需要保存起来'
        
        xt[partial_mask] = target[partial_mask]   # 最后输出时 观测处该为给定的观测值 
        
        
        samples = np.row_stack([samples, xt.detach().cpu().numpy()])
        reals = np.row_stack([reals, x.detach().cpu().numpy()])
        masks = np.row_stack([masks, t_m.detach().cpu().numpy()])
        
```
